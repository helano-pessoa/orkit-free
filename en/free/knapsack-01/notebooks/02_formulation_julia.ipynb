# 0-1 Knapsack Problem — Formulation and Solution (Julia)> **ORKit Free** · Integer Programming (IP)---## Formulation$$\max \sum_{i} v_i x_i \quad \text{s.t.} \quad \sum_{i} w_i x_i \leq W, \;\; x_i \in \{0,1\}$$

```julia
using JuMP, HiGHS, JSON3
```

## 1. Load Instance

```julia
data = JSON3.read(read("../instances/small_5.json", String))items = data.itemsW = data.capacityn = length(items)println("Capacity: $W, Items: $n")
```

## 2. Build and Solve

```julia
v = [it.value for it in items]w = [it.weight for it in items]model = Model(HiGHS.Optimizer)set_silent(model)@variable(model, x[1:n], Bin)@objective(model, Max, sum(v[i] * x[i] for i in 1:n))@constraint(model, sum(w[i] * x[i] for i in 1:n) <= W)optimize!(model)println("Status: ", termination_status(model))println("Value : ", objective_value(model))
```

## Key Takeaways- JuMP + HiGHS solves the 0-1 Knapsack efficiently via branch-and-bound- The binary constraint `Bin` models the take-it-or-leave-it decision- LP relaxation provides an upper bound for branch-and-bound
